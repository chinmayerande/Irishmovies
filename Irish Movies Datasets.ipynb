```python
### Now we want to gather more information about them by using a different url
```

```python
details="https://api.themoviedb.org/3/movie/{}?{}"
```

```python
json_list=[]
for title in list_of_titles:
    url=details.format(title,api_key)
    request=requests.get(url)
    if request.status_code!=200:
        continue
    else:
        data=request.json()
        json_list.append(data)
df=pd.DataFrame(json_list)
```

## Making a subset of df with important columns

```python
movies=df[["id","genres","original_title","release_date","runtime","spoken_languages","production_companies","production_countries","vote_average","vote_count"]]
```

```python
movies.genres
```

```python
movies.genres=movies.genres.apply(lambda x: "|".join(i["name"] for i in x) if isinstance(x,list) else np.nan)
```

```python
movies.spoken_languages[0]
```

```python
movies.spoken_languages=movies.spoken_languages.apply(lambda x: "|".join(i["english_name"] for i in x) if isinstance(x,list) else np.nan)
```

```python
movies["production_companies"]=movies.production_companies.apply(lambda x: "|".join(i["name"] for i in x) if isinstance(x,list) else np.nan)
```

```python
movies.production_countries=movies.production_countries.apply(lambda x: "|".join(i["name"] for i in x) if isinstance(x,list) else np.nan)
```
